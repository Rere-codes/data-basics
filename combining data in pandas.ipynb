```python
import pandas as pd

product=pd.DataFrame({
    'product_id':[101,102,103,104,105,106,107],
    'product_name':['Watch','Bag','Shoes','Smartphone','Books','Oil','Laptop'],
    'category':['Fashion','Fashion','Fashion','Electronics','Study','Grocery','Electronics'],
    'price':[299.0,1350.50,2999.0,14999.0,145.0,110.0,79999.0],
})

customer=pd.DataFrame({
    'customer_id':[1,2,3,4,5,6,7,8,9],
    'name':['David','Temitope','Kelicha','Ayomide','Garba','Divine','Lawal','Ajoke','Angel'],
    'age':[20,25,15,10,30,65,35,18,23],
    'product_id':[101,0,106,0,103,104,0,0,107],
    'purchased_product':['Watch','NA','Oil','NA','Shoes','Smartphone','NA','NA','Laptop'],
    'city_of_residence':['Yaba','Abeokuta','Ikorodu','Ibadan','Ibadan','Abeokuta','Lekki','Abeokuta','Ikeja']
})

```

```python
product
```

```python
customer
```

```python
product_customer = product.merge(customer, on='product_id')
```

```python
product_customer
```

```python
product_customer_1 = product.merge(customer, left_on='product_name', right_on='purchased_product')
```

```python
product_customer_1
```

```python
product_customer_2 = product.merge(customer, left_on='product_name', right_on='purchased_product', suffixes=('_prd','_cst'))
```

```python
product_customer_2
```

```python
left_prod_cust = product.merge(customer, on='product_id', how='left')
```

```python
left_prod_cust
```

```python
right_prod_cust = product.merge(customer, on='product_id', how='right')
```

```python
right_prod_cust
```

```python
out_prod_cust = product.merge(customer, on='product_id', how='outer')
```

```python
out_prod_cust
```

```python
out_prod_cust_ind = product.merge(customer, on='product_id', how='outer', indicator=True)
```
